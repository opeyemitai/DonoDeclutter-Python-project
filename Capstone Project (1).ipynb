```python
# Calculate frequency of each response option
response_counts = corrected['What do you usually do with items you no longer need? (Select all that apply)'].value_counts()
response_counts 
```

```python
# Plot the distribution of responses
plt.figure(figsize=(12, 10))
sns.barplot(x=response_counts.index, y=response_counts.values, palette='pastel')
plt.title('Distribution of Responses for Item Disposal')
plt.xlabel('Response')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

```python
# The column containing disposal method responses
disposal_methods = corrected['What do you usually do with items you no longer need? (Select all that apply)']

# Initialize an empty dictionary to store the counts of each disposal method
method_counts = {}

# Iterate over each response to count the frequency of disposal methods
for response in disposal_methods:
    # We will split the response if it contains multiple methods
    methods = response.split(', ')
    # Update the counts for each disposal method
    for method in methods:
        method = method.strip()  
        if method in method_counts:
            method_counts[method] += 1
        else:
            method_counts[method] = 1

# Converting the dictionary to a pandas Series for easier manipulation and visualization
method_counts_series = pd.Series(method_counts)

# Sort the Series by frequency in descending order
method_counts_series = method_counts_series.sort_values(ascending=False)
```

```python
# Plot the distribution of disposal methods using a pie chart
plt.figure(figsize=(10, 8))
plt.pie(method_counts_series.values, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), textprops={'fontsize': 12})
plt.title('Distribution of Reoccuring Disposal Methods for Unwanted Items')
plt.axis('equal')  
plt.legend(method_counts_series.index, loc='upper right', fontsize=10)
plt.tight_layout()
plt.show()
```

In the pie chart, 35.2% of respondents indicated that their disposal method for items was to give them to friends, while 12.8% preferred to sell them. Interestingly, approximately 29.1% of respondents expressed the importance of our decluttering application by choosing to dispose of items using its features.

## Reasons for Decluttering Analysis:

```python
# The column containing reasons for decluttering
disposal_methods = corrected['Reasons for decluttering']

# Initialize an empty dictionary to store the counts of each disposal method
method_counts = {}

# Iterate over each response to count the frequency of disposal methods
for response in disposal_methods:
    # We will split the response if it contains multiple methods
    methods = response.split(', ')
    # Update the counts for each disposal method
    for method in methods:
        method = method.strip()  
        if method in method_counts:
            method_counts[method] += 1
        else:
            method_counts[method] = 1

method_counts_series = pd.Series(method_counts)
method_counts_series = method_counts_series.sort_values(ascending=False)
```

```python
method_counts_series.unique()
```

```python
# Setting a threshold for considering reasons as "Other"
threshold = 0.05  #consider reasons with less than 5% frequency as "Other"
total_count = method_counts_series.sum()

# Filter reasons that meet the threshold
common_reasons = method_counts_series[method_counts_series / total_count > threshold]
other_count = total_count - common_reasons.sum()

# Append the "Other" category
common_reasons['Other'] = other_count

# Plot the distribution of reasons for decluttering using a pie chart
plt.figure(figsize=(10, 8))
plt.pie(common_reasons, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'), textprops={'fontsize': 12})
plt.title('Reasons for Decluttering')
plt.axis('equal')
plt.legend(common_reasons.index, loc='upper right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total count of reasons
total_count = method_counts_series.sum()

# Calculate the percentage of each motivation
percentage = (method_counts_series / total_count) * 100

# Sort the motivations by count in descending order
sorted_motivations = method_counts_series.sort_values(ascending=False)

# Select the top motivations (e.g., top 5)
top_motivations = sorted_motivations.head(5)

# Calculate the percentage of the top motivations
top_percentage = (top_motivations / total_count) * 100

print("Top motivations and their percentage among respondents:")
print(top_percentage)
```

In the analysis of reasons for decluttering, it was found that 33.8% of respondents cited seasonal cleaning as their primary motivation. Additionally, an equal proportion of 33.8% indicated the need for more space at home as their reason for decluttering. Furthermore, 12.0% of respondents stated that they were preparing to move or migrate, underscoring the various motivations behind decluttering practices.
